```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Padreganda-2012-SAT-Results.csv")
df.head()
```

```python
score_order = ['Low', 'Medium', 'High']
df['SAT Critical Reading Avg. Score'] = pd.Categorical(df['SAT Critical Reading Avg. Score'], categories=score_order, ordered=True)
df['SAT Math Avg. Score'] = pd.Categorical(df['SAT Math Avg. Score'], categories=score_order, ordered=True)
df['SAT Writing Avg. Score'] = pd.Categorical(df['SAT Writing Avg. Score'], categories=score_order, ordered=True)
```

```python
sns.set(style="whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Updated plotting with hue and legend=False
sns.countplot(data=df, x='SAT Critical Reading Avg. Score', hue='SAT Critical Reading Avg. Score',
              order=score_order, ax=axes[0], palette="Blues", legend=False)
axes[0].set_title('Distribution of SAT Reading Scores')

sns.countplot(data=df, x='SAT Math Avg. Score', hue='SAT Math Avg. Score',
              order=score_order, ax=axes[1], palette="Greens", legend=False)
axes[1].set_title('Distribution of SAT Math Scores')

sns.countplot(data=df, x='SAT Writing Avg. Score', hue='SAT Writing Avg. Score',
              order=score_order, ax=axes[2], palette="Reds", legend=False)
axes[2].set_title('Distribution of SAT Writing Scores')

plt.tight_layout()
plt.show()
```
